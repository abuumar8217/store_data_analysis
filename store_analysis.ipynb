**Labeling Age categories**

'Gen Z (0-24)': Ages from 0 to 24 (inclusive)

'Gen Y (25-39)': Ages from 25 to 39 (inclusive)

'Senior (60+)': Ages 60 and above

```python
import pandas as pd
import numpy as np
from datetime import datetime


df = pd.read_csv('SuperStore-Data.csv')
df.head()
```

```python
  df['Age']=datetime.now().year-df['Year_Birth']
  df.head()
```

```python
df.Income.fillna(df.Income.mean(), inplace=True)
```

```python

bins = [0, 25, 40, 55, 60, df.Age.max()+1]
labels = ['Gen Z', 'Gen Y', 'Gen X', 'Older', 'Senior']

# Create age category column
df['AgeCat'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df
```

```python
df.isna().sum()
```

```python
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
                'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

avg_by_age = df.groupby('AgeCat')[product_cols].mean().reset_index()

print(avg_by_age)

df_melted = avg_by_age.melt(id_vars='AgeCat',
                            var_name='Product',
                            value_name='AvgSpend')

print(df_melted)
```

```python
import plotly.express as px

fig = px.bar(
    df_melted,
    x='Product',
    y='AvgSpend',
    color='AgeCat',
    barmode='group',
    title='Average Spending by Age Category and Product Type',
    labels={'AvgSpend': 'Avg Monthly Spend ($)'},
    hover_data={'AvgSpend': ':.2f'}
)

fig.update_layout(template='plotly_dark', xaxis_tickangle=-45)
fig.show()
```

```python
# Edu vs Ave.Spending


# total spending
df['TotalSpent'] = df[['MntWines', 'MntFruits', 'MntMeatProducts',
                       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# ave spending vs Education
avg_spending_education = df.groupby('Education')['TotalSpent'].mean().reset_index()

print(avg_spending_education)


# ave spending vs Income
avg_spending_income = df.groupby('Income')['TotalSpent'].mean().reset_index()

print(avg_spending_income)
```

```python
fig1 = px.bar(
    avg_spending_education,
    x='Education',
    y='TotalSpent',
    color='Education',
    title='Education vs Average Spending',
    labels={'TotalSpent': 'Average Spending ($)', 'Education': 'Education Level'},
    hover_data=['Education', 'TotalSpent']
)


fig1.show()
```

```python
# Income vs Ave.Spending

# income bins
bins = [0, 20000, 50000, 100000, 150000, 200000, df.Income.max()+1]
labels = ['0-20K', '20K-50K', '50K-100K', '100K-150K', '150K-200K', '200K+']
df['IncomeBin'] = pd.cut(df['Income'], bins=bins, labels=labels, right=False)

# ave spending per bin
avg_spending_income_bin = df.groupby('IncomeBin')['TotalSpent'].mean().reset_index()
avg_spending_income_bin
```

```python

# bar chart
fig2 = px.bar(
    avg_spending_income_bin,
    x='IncomeBin',
    y='TotalSpent',
    title='Income Bins vs Average Spending',
    labels={'TotalSpent': 'Average Spending ($)', 'IncomeBin': 'Income Range'},
    color='IncomeBin',
    text='TotalSpent'

    )
fig2.update_traces(texttemplate='%{text:.2f}', hoverinfo='skip'
)
fig2.show()
```

```python
#  purchases across 3 channels (assuming catalog purchases are apart from web purchases)
purchase_data = {
    'Purchase Channel': ['Store Purchases', 'Web Purchases', 'Catalog Purchases'],
    'Total Purchases': [
        df['NumStorePurchases'].sum(),
        df['NumWebPurchases'].sum(),
        df['NumCatalogPurchases'].sum()
    ]
}

purchase_df = pd.DataFrame(purchase_data)

purchase_df['Percentage'] = (purchase_df['Total Purchases'] / purchase_df['Total Purchases'].sum()) * 100

purchase_df['Label'] = purchase_df.apply(lambda row: f"({row['Percentage']:.2f}%)", axis=1)

purchase_df
```

```python

fig = px.pie(
    purchase_df,
    names='Purchase Channel',
    values='Total Purchases',
    title='Market Distribution Across Purchase Channels',
    color='Purchase Channel',
    color_discrete_map={'Store Purchases': 'blue', 'Web Purchases': 'green', 'Catalog Purchases': 'orange'},
    hole=0.4,
    #labels={'Total Purchases': 'Number of Purchases'},
    #hover_data=['Label']
)

fig.update_traces(textinfo='percent+label', pull=[0, 0, 0.1])

fig.show()
```
